## Recursive search path

```python
def path_search(path):
    glob_search_path = os.path.join(path, "**/*.pcd")
    pcd_paths = glob2.glob(glob_search_path)
    print("pcd_paths", len(pcd_paths))
    return pcd_paths
    
```

## Simple visualization

```python
def simple_scatter(path):
    pcd_paths = path_search(path)
    
    for i in range(len(pcd_paths)):
        print(i)
        plt.clf
        p = pcl.load(pcd_paths[i])
        parray = np.asarray(p)
        frame = plt.scatter(parray[:,1],parray[:,0], parray[:,2], c="g", alpha=0.2)
        plt.show()
        plt.close()
```

```python
simple_scatter(dataset_path)
```

## PyntCloud Visualization

```python
def pyc_visu(path):
    pcd_paths = path_search(path)

    # i = randint(0, len(pcd_paths))
    for i in range(len(pcd_paths)):
        pcd_pyntcloud = PyntCloud.from_file(pcd_paths[i])
        #print(pcd_pyntcloud.points)
        #print(pcd_pyntcloud)
        scene = pcd_pyntcloud.plot(return_scene=True)
    
```

```python
pyc_visu(dataset_path)
```

## Ipyvolume visualization

```python
def ipy_visu(path):
    pcd_paths = path_search(path)
    
    #we are only going to create one image to avoid slowing down the notebook
    i = randint(0, len(pcd_paths))
    # for i in range(len(pcd_paths)):
    print(i)
    plt.clf
    p = pcl.load(pcd_paths[i])
    parray = np.asarray(p)
    # print(parray[:,1],parray[:,0], parray[:,2])
    x,y,z = parray[:,1],parray[:,0], parray[:,2]
    ipv.quickscatter(x,y,z, marker="sphere", size = 0.7)
    ipv.show()
```

```python
ipy_visu(dataset_path)
```
