```python
import dsdb_cli
```

# Filter PCDs.

```python
results = dsdb_cli.execute_command_filterpcds(
    number_of_points_threshold=37600, 
    confidence_avg_threshold=0.99, 
    remove_errors=True, 
    remove_rejects=True)

print("Number of results:", len(results["results"]))
results["results"][0:10]
```

```python
results = dsdb_cli.execute_command_filterpcds(
    sort_key="number_of_points", 
    sort_reverse=True)

print("Number of results:", len(results["results"]))
results["results"][0:10]
```

```python
results = dsdb_cli.execute_command_filterpcds(
    sort_key="number_of_points", 
    sort_reverse=False)

print("Number of results:", len(results["results"]))
results["results"][0:10]
```

# Sort PCDs.

```python
results = dsdb_cli.execute_command_sortpcds(
    sort_key="number_of_points", 
    sort_reverse=True)

print("Number of results:", len(results["results"]))
results["results"][0:10]
```

```python
results = dsdb_cli.execute_command_sortpcds(
    sort_key="number_of_points", 
    sort_reverse=False)

print("Number of results:", len(results["results"]))
results["results"][0:10]
```
